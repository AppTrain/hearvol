```r
loan_data = read.csv('../loan_data.csv')
```

```r
loan_data$outcome = ordered(loan_data$outcome, levels=c('paid off', 'default'))
```

```r
## Naive Bayes
naive_model = NaiveBayes(outcome ~ purpose_ + home_ + emp_len_, data = na.omit(loan_data))
```

```r
naive_model$table
```

```r
loan_data[147, c('purpose_', 'home_', 'emp_len_')]
```

```r
predict(naive_model,loan_data[147, c('purpose_', 'home_', 'emp_len_')])
```

```r
less_naive = NaiveBayes(outcome ~ borrower_score + payment_inc_ratio + purpose_ + home_ + emp_len_, data = loan_data)
```

```r
predict(less_naive,loan_data[147, c( 'borrower_score' , 'payment_inc_ratio','purpose_', 'home_', 'emp_len_')])
```

```r
less_naive$table
```

```r
library(ggplot2)
```

```r
stats = less_naive$table[[1]]
ggplot(data.frame(borrower_score=c(0,1)), aes(borrower_score)) +
  stat_function(fun = dnorm, color='blue', linetype=1, 
                arg=list(mean=stats[1, 1], sd=stats[1, 2])) +
  stat_function(fun = dnorm, color='red', linetype=2, 
                arg=list(mean=stats[2, 1], sd=stats[2, 2])) +
  labs(y='probability')
```
